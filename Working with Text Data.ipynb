## The expand and n Parameters of the .str.split() Method

```python
import pandas as pd
chicago = pd.read_csv('chicago.csv').dropna(how = 'all')
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail(3)
```

```python
chicago['Name'].str.split(',', expand = True).head()
```

```python
chicago[["First Name", "Last Name"]] = chicago['Name'].str.split(',',expand = True)
```

```python
chicago.head()
```

```python
chicago['Position Title'].str.split(' ', expand = True).head()
```

```python
chicago[['First Title Word', 'Remaining Words']] = chicago['Position Title'].str.split(' ', expand = True, n = 1).head()
```
